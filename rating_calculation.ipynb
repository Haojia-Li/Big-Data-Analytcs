```python
df1912 = pd.read_csv("D:/Fordham Spring Homework Assignment/Big Data Analytics (Yilu Zhou)/project presentation due May 8/data/2019-Dec.csv")
df1911 = pd.read_csv("D:/Fordham Spring Homework Assignment/Big Data Analytics (Yilu Zhou)/project presentation due May 8/data/2019-Nov.csv")
df1910 = pd.read_csv("D:/Fordham Spring Homework Assignment/Big Data Analytics (Yilu Zhou)/project presentation due May 8/data/2019-Oct.csv")
df2002 = pd.read_csv("D:/Fordham Spring Homework Assignment/Big Data Analytics (Yilu Zhou)/project presentation due May 8/data/2020-Feb.csv")
df2001 = pd.read_csv("D:/Fordham Spring Homework Assignment/Big Data Analytics (Yilu Zhou)/project presentation due May 8/data/2020-Jan.csv")
```

```python
df_list = [df1910,df1911,df1912,df2001,df2002]
df = pd.concat(df_list)
```

```python
len(df)
```

```python
df1 = df.copy()
```

```python
df1["count"] = 1 
```

```python
df1.head()
```

```python
event_count_df = df1.groupby([df1["user_id"], df1["product_id"], df1["event_type"]],as_index=False)["count"].sum()
```

```python
event_count_df
```

```python
event_count_df = event_count_df.pivot_table(index=['user_id','product_id'], columns='event_type', values='count')
```

```python
event_count_df.head()
```

```python
event_count_df = event_count_df.reset_index()
```

```python
event_count_df.head()
```

```python
event_count_df = event_count_df.fillna(0)
```

```python
event_count_df.head()
```

```python
rating = list()
cart = event_count_df["cart"]
purchase = event_count_df["purchase"]
remove = event_count_df["remove_from_cart"]
view = event_count_df["view"]
for i in range(0,len(event_count_df)):
    if purchase[i] + remove[i] > cart[i]:
        if remove[i] > cart[i]:
            temp_rating = (purchase[i]*5) + (view[i]*1)
        else:
            temp_rating = (cart[i]*2) + (purchase[i]*5) - (remove[i]*2) + (view[i]*1) + ((purchase[i]+remove[i]-cart[i])*2)
    else:
        temp_rating = (cart[i]*2) + (purchase[i]*5) - (remove[i]*2) + (view[i]*1)
    rating.append(temp_rating)
rating
```

```python
event_count_df["rating"] = rating
```
